```python
import requests as req
import pandas as pd
from datetime import datetime
```

```python
product = input('کالایی که میخای سرچ کن')

url = f'https://api.digikala.com/v1/search/?q={product}'

product_list = []
total_page = int(input('تعداد صفحه رو وارد کن'))

for page in range (0,total_page):
    page_url = f'{url}&page={page}'
    response = req.get(page_url)
    if response.status_code == 200:
        data=response.json()
        products = data['data']['products']

        for i in products:
            name = i['title_fa']
            price = i['default_variant']['price']['selling_price']
            slug = i['url']['uri']
            product_link = f'https://www.digikala.com{slug}'
            today_date = datetime.now().strftime('%Y-%m-%d')

            product_list.append(
                {
                    'id':i['id'],
                    'name':name,
                    'price_toman':price,
                    'link':product_link,
                    'date':today_date,

                }
            )

    else:
        print('خطا در دریافت اطلاعات')
```

```python
dataset = pd.DataFrame(product_list)
dataset.to_csv('digikala_dataset.csv',index=False)
```
